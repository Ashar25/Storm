```python
import numpy as np
import pandas as pd
n_years = 1
df_gauge_tot = pd.read_excel(r'D:\IISc\Project\Stormbang\GaugeTot_{}y.xlsx'.format(n_years),index_col=0)
df_gauge_int = pd.read_excel(r'D:\IISc\Project\Stormbang\Gaugeint_{}y.xlsx'.format(n_years),index_col=0)
df_gauge_tot
```

```python
num_gauges = 554
x = np.zeros((num_gauges,4))
s = []
for i in range(0,num_gauges):
    df_new_tot = df_gauge_tot.loc[:,i].copy()
    df_new_int = df_gauge_int.loc[:,i].copy()
    df_new1 = df_new_tot[df_new_tot != 0]
    ptot = df_new_tot.values
    pint = df_new_int.values
    
    y = df_new1.shape[0]
    som = np.sum(ptot)/y
    som1 = np.sum(pint)/y
    x[i,0] = y;x[i,1] = som;x[i,2]=som1;x[i,3] = 60* som/som1;
#     x.append(y)
#     s.append(som)
df_gaugeshit = pd.DataFrame(x,columns = ['Number of storms hit in' + str(n_years)+ 'years','Average Depth (mm)','Average Intensity(mm/hr)'
                                        ,'Average Duration (min)'],)
# dfsum = pd.DataFrame(s)
# dfsum
df_gaugeshit.to_csv('Average gauges hit{}.csv'.format(n_years))
```
